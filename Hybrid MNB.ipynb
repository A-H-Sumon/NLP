```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
```

```python
stop_words = set(stopwords.words("english"))
```

```python
from nltk.stem.porter import PorterStemmer 
from nltk.stem import WordNetLemmatizer
```

```python
wordnet = WordNetLemmatizer()
def text_preproc(x):
    x = x.lower()
    x = ' '.join([word for word in x.split( ) if word not in stop_words])
    x = x.encode('ascii', 'ignore').decode()
    x = re.sub(r'https*\S+', ' ', x)
    x = re.sub(r'@\S+', ' ', x)
    x = re.sub(r'#\S+', ' ', x)
    x = re.sub(r'\'\w+', '', x)
    x = re.sub('[%s]' % re.escape(string.punctuation), ' ', x)
    x = re.sub(r'\w*\d+\w*', '', x)
    x = re.sub(r'\s{2,}', ' ', x)
    x = re.sub(r'\[]', '', x)
    x = re.sub(r'\"', '', x)
    x = re.sub(r'digits', ' ',x)
    return x
```

```python
positive_pattern1 = r"happy"
positive_pattern2 = r"enjoy"

positive_pattern5 = r"love"
positive_pattern6 = r"enjoyable"
positive_pattern7 = r"exciting"


positive_pattern_list = [positive_pattern1,positive_pattern2,positive_pattern5,positive_pattern6,positive_pattern7]
```

```python
positive_complex_pattern = re.compile('|'.join(['(%s)' % i for i in positive_pattern_list]),re.IGNORECASE)
```

```python
negative_pattern1 = r"die"
negative_pattern2 = r"suicide"
negative_pattern3 = r"kill"
negative_pattern5 = r"pain"
negative_pattern6 = r"depression"
negative_pattern7 = r"suicidal"
negative_pattern9 = r"worse"
negative_pattern10 = r"lost"
negative_pattern12 = r"death"
negative_pattern13 = r"scared"

negative_pattern_list = [negative_pattern1,negative_pattern2,negative_pattern3,negative_pattern5,negative_pattern6,negative_pattern7,negative_pattern9,negative_pattern10,negative_pattern12,negative_pattern13]
```

```python
negative_complex_pattern = re.compile('|'.join(['(%s)' % i for i in negative_pattern_list]),re.IGNORECASE)
```

```python
df["negative_sentiment_flag"] = df["clean_text"].apply(lambda x:1 if(len(re.findall(negative_complex_pattern,x))>0) else 0)
```

```python
df["positive_sentiment_flag"] = df["clean_text"].apply(lambda x:1 if(len(re.findall(positive_complex_pattern,x))>0) else 0)
```
